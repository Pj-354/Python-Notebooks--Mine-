```python
xls = pd.ExcelFile(r'Data3.xlsx')
df_returns = pd.read_excel(xls, 'returns', index_col=0)
df_returns.columns = pd.to_datetime(df_returns.columns, format = '%Y%m').to_period('M')
df_EPS = pd.read_excel(xls, 'EPS', index_col=0)
df_EPS.columns = pd.to_datetime(df_EPS.columns, format = '%Y%m').to_period('Q-APR')
df_consensus = pd.read_excel(xls, 'consensus', index_col= 0)
df_consensus.columns = pd.to_datetime(df_consensus.columns, format='%Y%m').to_period('Q-APR')
```

```python
df_returns_quarterly = df_returns.T.groupby(pd.PeriodIndex(df_returns.columns, freq='Q-APR')).mean()
df_returns_quarterly.loc['2020Q2':'2021Q1'].T
```

```python
#EPS - consensus for each reporting month
delta_EPS = df_EPS - df_consensus
delta_EPS_2016_2019 = df_EPS.loc[:, '2016Q1':'2019Q4'] - df_consensus.loc[:, '2016Q1':'2019Q4']
#Standard Deviation of (EPS - Consensus)
stdev_EPS_estimates = np.array((delta_EPS_2016_2019).std(axis=1))
#SUE of all Stocks
df_SUE = delta_EPS.apply(lambda col : col / stdev_EPS_estimates, axis = 0)
df_SUE
```

```python
''' Selection Criteria for Stocks:
        1. Rank all stocks based on SUE
        2. Select top and bottom N = 30 stocks
        3. Sample K = 10 randomly from each criteria
        4. Repeat for every announcement period
 '''



def SUE_construction(df_SUE = df_SUE, N = 30, K= 10, Top_K = True,
                      df_returns = df_returns, seed = 2007,
                      start = '2020-03', end = '2022-02'):

    df_returns = df_returns.loc[:, start:end]
    df_returns_quarterly = df_returns.T.groupby(pd.PeriodIndex(df_returns.columns, freq='Q')).mean()
    df_returns_quarterly = df_returns_quarterly.loc[start:end]
    df_returns_quarterly = df_returns_quarterly.T

    pr = pd.period_range(pd.Period(start, freq='Q'), pd.Period(end, freq='Q'), freq='Q-APR')
    df_SUE = df_SUE[pr]

    def top_SUE(df_SUE = df_SUE, N = N, K = K, seed = seed, df_returns = df_returns):

        good_news_stocks = {} #EPS of stocks selected that are 'good news stocks'
        tickers = {}        
        np.random.seed(seed)
        for i in range(0, len(df_SUE.columns)):
            #Samples from N stocks to get K stocks per period
            good_news_stocks[i] = df_SUE.iloc[:,i].nlargest(N).sample(K)

            #Collects names of stocks in our portfolio
            tickers[i] = list(good_news_stocks[i].index)

        #DataFrame contains portfolio at each period
        tickers = pd.DataFrame(tickers)
        tickers.columns = df_SUE.columns

        good_news_stocks = pd.DataFrame(good_news_stocks)
        return tickers, good_news_stocks

    def bottom_SUE(df_SUE = df_SUE, N = N, K = K, seed = seed, df_returns = df_returns):
    
        bad_news_stocks = {} #EPS of stocks selected that are 'good news stocks'
        tickers = {}        
        np.random.seed(seed)
        for i in range(0, len(df_SUE.columns)):
            #Samples from N stocks to get K stocks per period
            bad_news_stocks[i] = df_SUE.iloc[:,i].nsmallest(N).sample(K)

            #Collects names of stocks in our portfolio
            tickers[i] = list(bad_news_stocks[i].index)

        #DataFrame contains portfolio at each period
        tickers = pd.DataFrame(tickers)
        tickers.columns = df_SUE.columns

        bad_news_stocks = pd.DataFrame(bad_news_stocks)
        return tickers, bad_news_stocks
    
    if Top_K == True:
        tickers, stocks = top_SUE()

    if Top_K == False:
        tickers, stocks = bottom_SUE()
    
    #Calculate Average Returns of Portfolios
    #Apply Boolean mask then take arithmetic mean
    portfolio_returns_mask = ~stocks.isna()
    portfolio_returns_mask.columns = df_returns_quarterly.columns

    returns_per_individual_stock = df_returns_quarterly[portfolio_returns_mask]
    portfolio_returns = pd.Series(returns_per_individual_stock.mean(axis='rows'))

    return tickers, returns_per_individual_stock, portfolio_returns, stocks
```

## Simulating SUE Strategy with N = 5, 20, 100, 1000 Trials

```python
#Sample Runs (Pseudo Random)
Simulations = 5
seed_array = np.arange(0, Simulations)

#Annualised returns function
def annualised_returns(r, periods = 1):
    compounded_growth = (1 + r).prod()
    return compounded_growth ** (periods / 4) - 1

#Running the Test
Good_portfolio_rets = []
Bad_portfolio_rets = []
Long_Short_portfolio_rets = []

for i in range(0, Simulations):
    #Pseudo-Random Run
    seed = seed_array[i]
    Good_portfolio_rets.append(annualised_returns(SUE_construction(seed = seed)[2]))
    Bad_portfolio_rets.append(annualised_returns(SUE_construction(Top_K = False, seed= seed)[2]))
    Long_Short_portfolio_rets.append(annualised_returns(Good_portfolio_rets[i] - Bad_portfolio_rets[i]))

Performance_df = pd.DataFrame({'Good' : np.mean(Good_portfolio_rets).round(4) * 100,
                              'Bad' : np.mean(Bad_portfolio_rets).round(4) * 100,
                              'Long/Short' : np.mean(Long_Short_portfolio_rets).round(4) * 100}
                              , index = [f'Average Annualised Returns after {Simulations} runs'])

```

# CAPM and FF-Model Regression 

```python

''' 
Apply regression to each column
    - SMF doesn't like spaces in columns
    - Function needs to take pd.Series object of column names
    - Lambda + Apply => pass through arguements as below
    - Output DF index name : column 'Portfolio' can be set as index using .set_index()

'''

def regression_CAPM(portfolio_col, factors, df = df_portfolio_excess, factor_df = df_FF2):                               
    
    ''' 
    Regresses columns in portfolio_col against the factors passed 
    
    '''
    #Slice portfolio and factor values for desired time frame
    returns = df[f'{portfolio_col}']
    factor_values = factor_df[f'{factors}']
    regression_df = pd.concat([returns, factor_values], axis='columns')
    #Create Regression Model

    model = smf.ols(f'{portfolio_col} ~ {factors}', data = regression_df)
    results = model.fit(use_T = True)
    
    return pd.Series(
            {
        'Portfolio' : f'{portfolio_col}',
        'Intercept (Alpha)' : results.params['Intercept'].round(4) * 100,
        'Intercept (T-Value)' : results.tvalues['Intercept'].round(3),
        'Market Risk' : results.params[f'{factors}'].round(4),
         'Market Risk (T-Value)' : results.tvalues[f'{factors}'].round(3)
            },
        )

column_names = pd.Series(df_portfolio_excess.columns)
CAPM_Regression_Table = column_names.apply(lambda col : regression_CAPM(col, factors = 'MktRf')).set_index('Portfolio')
CAPM_Regression_Table
```

```python

def regression_FF(portfolio_col, factors, df = df_portfolio_excess, factor_df = df_FF2):
    
    ''' 
    Regresses columns in portfolio_col against the factors passed 
    
    '''
    returns = df[f'{portfolio_col}']
    factor_values = factor_df[factors]
    regression_df = pd.concat([returns, factor_values], axis='columns')

    #Create Regression Model
    model = smf.ols(f'{portfolio_col} ~ {factors[0]} + {factors[1]} + {factors[2]}', data = regression_df)
    results = model.fit(use_T = True)
    
    return pd.Series(
            {
        'Portfolio' : f'{portfolio_col}',
        'Intercept' : results.params['Intercept'].round(4) * 100,
        'Intercept (T-Value)' : results.tvalues['Intercept'].round(3),
        f'{factors[0]}' : results.params[f'{factors[0]}'].round(3),
        f'{factors[0]} T-Value' : results.tvalues[f'{factors[0]}'].round(3),
        f'{factors[1]}' : results.params[f'{factors[1]}'].round(3),
        f'{factors[1]} T-Value' : results.tvalues[f'{factors[1]}'].round(3),
        f'{factors[2]}' : results.params[f'{factors[2]}'].round(3),
        f'{factors[2]} T-Value' : results.tvalues[f'{factors[2]}'].round(3)
            },
        )

FF_Regression = column_names.apply(lambda col : regression_FF(col, factors=['MktRf', 'SMB', 'HML'])).set_index('Portfolio')

FF_Regression
```
