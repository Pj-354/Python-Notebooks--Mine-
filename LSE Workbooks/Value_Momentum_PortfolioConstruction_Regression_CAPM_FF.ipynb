# Question 1: Size Effect

Portfolio Requirements:

Small/large constitutes top (or bottom) 30 stocks ~ 10%
    2. 10 stocks will be selected for each portfolio - randomly sampled from top 50
    2. Held for 1 month
    3. Rebalanced every month
    4. Equally weighted

```python

def portfolio_returns(top_K, df_by, N = 10, K= 30, returns_df = df_returns, seed = 406): 
    '''  
    Function will order all the firms by e.g., size/value for each time period
        - Will then select the top 30 in that month
        - Sample (psuedo-) randomly to get 10 stocks
        - Compile into 1 portfolio, take average returns for that month
        - Repeat 
    '''

    def largest_values(df_by, returns_df, seed, N, K ):

        portfolio_factor_values = {}
        tickers = {}        
        np.random.seed(seed)

        for i in range(0, len(returns_df.columns)):
            #Samples from K stocks to get N stocks per period
            portfolio_factor_values[i] = df_by.iloc[:,i].nlargest(K).sample(N)

            #Collects names of stocks in our portfolio
            tickers[i] = list(portfolio_factor_values[i].index)

        #DataFrame contains portfolio to be held
        tickers = pd.DataFrame(tickers)
        tickers.columns = returns_df.columns

        #DataFrame contains value of factors in question e.g., market caps of top 10 largest
        portfolio_factor_values = pd.DataFrame(portfolio_factor_values)
        portfolio_factor_values.columns = returns_df.columns

        return tickers, portfolio_factor_values

    def lowest_values(df_by, returns_df, seed, N, K ):
        portfolio_factor_values = {}
        tickers = {}        
        np.random.seed(seed)
        for i in range(0, len(returns_df.columns)):
            portfolio_factor_values[i] = df_by.iloc[:,i].nsmallest(K).sample(N)
            tickers[i] = list(portfolio_factor_values[i].index)
        tickers = pd.DataFrame(tickers)
        tickers.columns = returns_df.columns
        portfolio_factor_values = pd.DataFrame(portfolio_factor_values)
        portfolio_factor_values.columns = returns_df.columns
        return tickers, portfolio_factor_values

    #Calculate Average Returns of Portfolios
    if top_K == True:
        tickers, portfolio_factor_values = largest_values(df_by, returns_df, seed, N, K)

    if top_K == False:
        tickers, portfolio_factor_values = lowest_values(df_by, returns_df, seed, N, K)

    #Apply Boolean mask then take arithmetic mean
    portfolio_factor_values_mask = ~portfolio_factor_values.isna()
    returns_per_individual_stock = returns_df[portfolio_factor_values_mask]
    portfolio_returns = pd.Series(returns_per_individual_stock.mean(axis='rows'))

    return tickers, returns_per_individual_stock, portfolio_returns 
```

```python
#Creating Portfolios for Large, Small Size Effect
tickers_large, returns_large_per_stock, returns_large = portfolio_returns(top_K= True, df_by = df_cap,
                                                                          returns_df = df_returns)

tickers_small, returns_small_per_stock, returns_small = portfolio_returns(top_K = False, df_by = df_cap,
                                                                          returns_df = df_returns)

#Creating Portfolios for Value Effect
tickers_value, returns_value_per_stock, returns_value = portfolio_returns(top_K = True, df_by= df_BM)
tickers_growth, returns_growth_per_stock, returns_growth = portfolio_returns(top_K = False, df_by= df_BM)
```

```python
''' 
Calculate annualised returns (as if they were an actual strategy) to factor in volatility drag

'''
def annualised_returns(r, periods):
    compounded_growth = (1 + r).prod()
    n = r.shape[0]
    return compounded_growth ** (periods / n) - 1

q1_small_returns = annualised_returns(returns_small, 12)
q1_large_returns = annualised_returns(returns_large, 12)
q1_smb_returns = annualised_returns(returns_small - returns_large, 12)

q1_value_returns = annualised_returns(returns_value, 12)
q1_growth_returns = annualised_returns(returns_growth, 12)
q1_value_effect_returns = annualised_returns(returns_value - returns_growth, 12)
```

Smaller stocks don't seem to perform as well as larger stocks when sampled randomly in each period. Potentially either rebalancing is too frequent (small firms can list and go under quickly), survivorship bias is not accounted for, size effect has been arb-ed out post 1990s, favourable macro for larger companies e.g., from 90s-2024 there has been COVID, dotcom, GFC, trade-wars, interest rates, tech trends that favour large companies (e.g., MAG7 outperformance)

```python
#Portfolio Returns into DF
df_portfolio = pd.concat([returns_growth, returns_value, returns_value - returns_growth, 
                          returns_small, returns_large, returns_small - returns_large]
                          ,axis = 'columns')
df_portfolio.index = pd.date_range(start='1990-01', periods=408, freq ='ME').to_period('M')
df_portfolio.columns = ['Growth', 'Value', 'LS_Value_Growth', 'Small', 'Big', 'LS_Small_Big']

#Portfolio of Returns into Excess Returns
df_portfolio_excess = df_portfolio.apply(lambda col : col - df_FF2['RF'])
df_portfolio_excess
```

```python

def regression_CAPM(portfolio_col, factors, df = df_portfolio_excess, factor_df = df_FF2
    ,timeframe = pd.period_range(pd.Period('1990-01', freq='M'),
                                pd.Period('2023-12', freq='M'))
                                ):
    
    ''' 
    Regresses columns in portfolio_col against the factors passed 
    
    '''
    #Slice portfolio and factor values for desired time frame
    returns = df[f'{portfolio_col}'].loc[timeframe]
    factor_values = factor_df[f'{factors}'].loc[timeframe]
    regression_df = pd.concat([returns, factor_values], axis='columns')
    #Create Regression Model

    model = smf.ols(f'{portfolio_col} ~ {factors}', data = regression_df)
    results = model.fit(use_T = True)
    
    return pd.Series(
            {
        'Portfolio' : f'{portfolio_col}',
        'Intercept' : results.params['Intercept'].round(3),
        'Intercept (T-Value)' : results.tvalues['Intercept'].round(3),
        'Market Risk' : results.params[f'{factors}'].round(3),
         'Market Risk (T-Value)' : results.tvalues[f'{factors}'].round(3)
            },
        )
```

```python

def regression_FF(portfolio_col, factors, df = df_portfolio_excess, factor_df = df_FF2
    ,timeframe = pd.period_range(pd.Period('1990-01', freq='M'),
                                pd.Period('2023-12', freq='M'))
                                ):
    
    ''' 
    Regresses columns in portfolio_col against the factors passed 
    
    '''
    #Slice portfolio and factor values for desired time frame
    returns = df[f'{portfolio_col}'].loc[timeframe]
    factor_values = factor_df[factors].loc[timeframe]
    regression_df = pd.concat([returns, factor_values], axis='columns')

    #Create Regression Model
    model = smf.ols(f'{portfolio_col} ~ {factors[0]} + {factors[1]} + {factors[2]}', data = regression_df)
    results = model.fit(use_T = True)
    
    return pd.Series(
            {
        'Portfolio' : f'{portfolio_col}',
        'Intercept' : results.params['Intercept'].round(3),
        'Intercept (T-Value)' : results.tvalues['Intercept'].round(3),
        f'{factors[0]}' : results.params[f'{factors[0]}'].round(3),
        f'{factors[0]} T-Value' : results.tvalues[f'{factors[0]}'].round(3),
        f'{factors[1]}' : results.params[f'{factors[1]}'].round(3),
        f'{factors[1]} T-Value' : results.tvalues[f'{factors[1]}'].round(3),
        f'{factors[2]}' : results.params[f'{factors[2]}'].round(3),
        f'{factors[2]} T-Value' : results.tvalues[f'{factors[2]}'].round(3)
            },
        )

regression_FF(portfolio_col = 'Growth', factors=['Market_Risk', 'SMB', 'HML'])

```

```python
''' 
Apply regression to each column
    - Function needs to take pd.Series object of column names
    - Lambda + Apply => pass through arguements as below
    - Output DF index name : column 'Portfolio' can be set as index using .set_index()

'''
portfolio_columns = pd.Series(df_portfolio_excess.columns)

results_df = portfolio_columns.apply(lambda col : 
                    regression_CAPM(portfolio_col = col, factors = 'Market_Risk'))


results_df2 = pd.DataFrame(results_df).set_index('Portfolio')

results_df3 = portfolio_columns.apply(
    lambda col : regression_FF(col, factors=['Market_Risk', 'SMB', 'HML'],
                               df = df_portfolio_excess, factor_df = df_FF2)
).set_index('Portfolio')
```
